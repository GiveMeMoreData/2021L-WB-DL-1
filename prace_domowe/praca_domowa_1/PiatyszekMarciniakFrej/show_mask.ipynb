```python
import numpy as np
import matplotlib.pylab as plt
import matplotlib.patches as mpatches
```

```python
db = np.load('/opt/notebooks/LungNet/lidc-val.npz', allow_pickle=True)['db'][()]
```

```python
obs = np.random.choice(len(db['X']))
X = db['X'][obs]
Y = db['Y'][obs]
```

```python
print(Y.shape)
print(X.shape)
```

```python
for i in range(3):
    obs = np.random.choice(len(db['X']))
    X = db['X'][obs]
    Y = db['Y'][obs]

    all_labels = ['nodule', 'background']
    colors = np.asarray([[255,100,100,155],
                         [255,255,255,0]]) / 255.
    Y_lbls = np.argmax(Y,axis=-1)

    plt.rcParams["figure.figsize"] = (20,10)
    plt.subplot(1, 2, 1)
    plt.imshow(np.squeeze(X),cmap='gray')
    plt.subplot(1, 2, 2)
    plt.imshow(np.squeeze(X),cmap='gray')
    plt.imshow(colors[np.squeeze(Y_lbls)])
    plt.legend(handles=[mpatches.Patch(color=colors[i], label=all_labels[i]) for i in np.unique(Y_lbls)], prop={'size':6})
    plt.show()
```
